```python
import pandas as pd
df1 = pd.read_csv("D:\\DATA SCIENCE\\DS INTERNSHIP\\HAMOYE INTERNSHIP\\Hamoye Fall'23 Cohort\\Hamoye Fall'23 Dataset\\FoodBalanceSheets_E_Africa_NOFLAG (1).csv", encoding = "latin-1")
```

```python
df1.shape
```

```python
import pandas as pd

df1['Element'] = df1['Element'].str.strip()

# Group the DataFrame by the 'Element' column
grouped_by_element = df1.groupby('Element')

# list of columns to sum
columns_to_sum = ['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']

# Calculating the sum of 'Stock Variation' for each year ('Y2014' to 'Y2018') within each group
sum_stock_variation = grouped_by_element[columns_to_sum].sum()

# Find the year with the highest sum of 'Stock Variation' for each element
highest_year = sum_stock_variation.idxmax(axis=1)

print("Year with the highest sum of Stock Variation for each element:")
print(highest_year)
```

```python
import pandas as pd


# To Select 'Y2017' and 'Area' columns
selected_columns = df1[['Y2017', 'Area']]

# Group by 'Area' and calculate the sum of 'Y2017' for each area
grouped_by_area = selected_columns.groupby('Area').sum()

# Sort the groups by the sum of 'Y2017' in ascending order
sorted_by_sum = grouped_by_area.sort_values(by='Y2017', ascending=True)

# Get the 7th group, which corresponds to the 7th lowest sum in 'Y2017'
seventh_lowest_area = sorted_by_sum.index[6]

print("Area with the 7th lowest sum in Y2017:", seventh_lowest_area)
```

```python
import pandas as pd


# Selecting 'Y2017' and 'Area' columns
selected_columns = df1[['Y2017', 'Area']]

# Group by 'Area' and calculate the sum of 'Y2017' for each area
grouped_by_area = selected_columns.groupby('Area').sum()

# Find the area with the highest sum in 'Y2017'
highest_sum_area = grouped_by_area['Y2017'].idxmax()

print("Area with the highest sum in Y2017:", highest_sum_area)

```

```python
wine_production = df1.groupby('Item')[['Y2015', 'Y2018']].sum()

# Print the total sum of Wine produced in 2015 and 2018
wine_2015_sum = wine_production.loc['Wine', 'Y2015']
wine_2018_sum = wine_production.loc['Wine', 'Y2018']

print("Total Wine Production in 2015:", wine_2015_sum)
print("Total Wine Production in 2018:", wine_2018_sum)
```

```python
# Calculate the correlation between 'Element Code' and each year
correlation_2014 = df1['Element Code'].corr(df1['Y2014'])
correlation_2015 = df1['Element Code'].corr(df1['Y2015'])
correlation_2016 = df1['Element Code'].corr(df1['Y2016'])
correlation_2017 = df1['Element Code'].corr(df1['Y2017'])
correlation_2018 = df1['Element Code'].corr(df1['Y2018'])

# Create a dictionary to store correlations for each year
correlations = {
    'Y2014': correlation_2014,
    'Y2015': correlation_2015,
    'Y2016': correlation_2016,
    'Y2017': correlation_2017,
    'Y2018': correlation_2018
}

# Find the year with the least correlation
least_correlation_year = min(correlations, key=correlations.get)

print("Year with the least correlation with 'Element Code':", least_correlation_year)
```
